```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel('UK Death Rate during Covid.xlsx')
df.head()
```

```python
df.info()
```

```python
df.duplicated()
```

```python
grouped_data = df.groupby(['Vaccination status', 'Age group']).sum()

vaccination_statuses = df['Vaccination status'].unique()

for status in vaccination_statuses:
    status_data = grouped_data.loc[status]
    plt.figure(figsize=(12, 6))
    plt.plot(status_data.index, status_data['Person-years'], marker='o', label='Person-years')
    plt.plot(status_data.index, status_data['Deaths involving COVID-19'], marker='o', label='Deaths involving COVID-19')
    plt.plot(status_data.index, status_data['Non-COVID-19 deaths'], marker='o', label='Non-COVID-19 deaths')
    plt.plot(status_data.index, status_data['All deaths'], marker='o', label='All deaths')

    plt.title(f'{status} - Vaccination Status Across Age Groups')
    plt.xlabel('Age Group')
    plt.ylabel('Count')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
import matplotlib.pyplot as plt

df = pd.read_excel('Monthly age-standardised mortality rates by vaccination status.xlsx')
df.head()
```

```python
df.info()
```

```python
df.duplicated()
```

```python
unvaccinated_data = df[df['Vaccination status'] == 'Unvaccinated']

plt.figure(figsize=(12, 6))
plt.plot(unvaccinated_data['Month'], unvaccinated_data['Number of deaths'], marker='o')
plt.title('Number of Deaths over Months for Unvaccinated')
plt.xlabel('Month')
plt.ylabel('Number of Deaths')
plt.grid(True)
plt.show()

plt.figure(figsize=(12, 6))
plt.bar(unvaccinated_data['Month'], unvaccinated_data['Age-standardised mortality rate per 100,000 person-years'])
plt.title('Age-standardised Mortality Rate over Months for Unvaccinated')
plt.xlabel('Month')
plt.ylabel('Age-standardised Mortality Rate per 100,000 person-years')
plt.grid(True)
plt.show()

plt.figure(figsize=(12, 6))
plt.scatter(unvaccinated_data['Person-years'], unvaccinated_data['Number of deaths'])
plt.title('Scatter Plot: Person-years vs Number of Deaths for Unvaccinated')
plt.xlabel('Person-years')
plt.ylabel('Number of Deaths')
plt.grid(True)
plt.show()
```

```python
# Load external data (replace 'your_data.csv' with your actual file path)
df = pd.read_excel('districts.xlsx')
df.head()
```

```python
df.info()
```

```python
df.duplicated()
```

```python
bar_data = df[df['Date'] == '26-04-2020'].groupby('State')['Confirmed'].sum().reset_index()
plt.figure(figsize=(15, 6))
plt.bar(bar_data['State'], bar_data['Confirmed'], color='skyblue')
plt.xlabel('State')
plt.ylabel('Confirmed Cases')
plt.title('Total Confirmed Cases by State on 26-04-2020')
plt.xticks(rotation=90)
plt.show()

line_data = df[df['State'] == 'Maharashtra'].groupby('Date')['Confirmed'].sum().reset_index()
plt.figure(figsize=(15, 6))
plt.plot(line_data['Date'], line_data['Confirmed'], marker='o', color='orange')
plt.xlabel('Date')
plt.ylabel('Confirmed Cases')
plt.title('Total Confirmed Cases Trend in Maharashtra')
plt.xticks(rotation=90)
plt.show()

pie_data = df[(df['State'] == 'Gujarat') & (df['Date'] == '26-04-2020')]
plt.figure(figsize=(15,15))
plt.pie(pie_data['Confirmed'], labels=pie_data['District'], autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Distribution of Confirmed Cases in Gujarat on 26-04-2020')
plt.show()
```
